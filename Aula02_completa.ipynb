Desafio: Crie um código que percorra a lista de médias e some 1 a cada valor, exceto quando o valor passar de 10.0, que aí ele deve parar em 10.0.

```python
n = 0

while n < 5:
    lista_de_medias[n] = lista_de_medias[n] + 1.0
    if lista_de_medias[n] > 10.0:
        lista_de_medias[n] = 10.0

    n = n + 1

print(lista_de_medias)
```

```python
from groq import Groq

client = Groq()
completion = client.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=[
      {
        "role": "user",
        "content": "DE QUE E FEITO O SOL?"
      }
    ],
    temperature=1, #0 o menos criativo possivel; #2 o mais criativo possivel.
    max_completion_tokens=8192,
    top_p=1,
    reasoning_effort="medium",
    stream=True,
    stop=None
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```

```python
lista_de_perguntas = [
    "como vou fazer para arrumar um emprego?",
    "por que e tao dificil?",
    "a culpa e minha?",
    "quando o  brasil vai melhorar?",
    "como estudar certo?",
    "sera que esse curso da escola da nuvem vai realmente me ajudar a arrumar um emprego?",
    "o que sera de mim se eu n conseguir?",
    "e as minhas dividas?",
    "e o meu futuro?",
    "o que fazer?"



]
with open("perguntas.txt", "w", encoding="utf-8") as arquivo:
  for pergunta in lista_de_perguntas:
    arquivo.write(pergunta + "\n")
```

```python
lista_desafio = []

with open("perguntas.txt", "r", encoding="utf-8") as arquivo:
  for linha in arquivo:
    lista_desafio.append(linha.strip())
print(lista_desafio)
```

```python
lista_de_dicionarios_de_respostas = []

for pergunta in lista_desafio:
    resposta = client.models.generate_content(
      model="gemini-2.5-flash",
      contents= f"Gere uma resposta muito sucinta para a pergunta:{pergunta}"
      )
    lista_de_dicionarios_de_respostas.append({"pergunta":pergunta, "resposta":resposta.text})
```

```python
lista_de_dicionarios_de_respostas
```

```python
with open("'respostas1.csv", "w", encoding="utf-8") as arquivo:
  arquivo.write("pergunta,resposta\n")
  for pergunta_dict in lista_de_dicionarios_de_respostas:
      arquivo.write(f"{pergunta_dict['pergunta']},{pergunta_dict['resposta']}\n")
```

```python
import pandas as pd
import numpy as np
import random

# Gerar 50 linhas de dados
num_rows = 50

# Nomes de produtos (exemplo)
product_names = [f'Produto {i+1}' for i in range(num_rows)]

# Categorias de produtos (exemplo)
categories = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros', 'Casa e Decoração']
product_categories = random.choices(categories, k=num_rows)

# Preço do produto (entre 10.00 e 500.00)
product_prices = np.round(np.random.uniform(10.00, 500.00, num_rows), 2)

# Quantidade de itens vendidos (entre 1 e 100)
items_sold = np.random.randint(1, 101, num_rows)

# Avaliação do produto (entre 1.0 e 5.0)
product_ratings = np.round(np.random.uniform(1.0, 5.0, num_rows), 1)

# Criar o DataFrame
df_produtos = pd.DataFrame({
    'nome do produto': product_names,
    'categoria do produto': product_categories,
    'preco do produto': product_prices,
    'quantos itens foram vendidos': items_sold,
    'avaliacão do produto': product_ratings
})

# Exibir as primeiras 5 linhas do DataFrame
display(df_produtos.head())
```
